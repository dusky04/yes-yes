### Pixel Intensity Sampling
Paper: `An Empirical Comparison of Video Frame Sampling Methods for Multi-Modal RAG Retrieval (https://arxiv.org/pdf/2408.03340v1)`

```python
from decord import VideoReader
import matplotlib.pyplot as plt
from math import sqrt
from PIL import Image
import torch
import seaborn as sns
import numpy as np
from torchvision.transforms import Resize
%matplotlib inline
```

```python
# get resolution of each frame
frame = vr[2].asnumpy()
frame = Image.fromarray(frame)
frame = frame.resize((224, 224))
frame = np.array(frame)
print("Frame Size: ", frame.shape)
# (224, 224, 3) -> (H, W, C)
```

```python
plt.figure(figsize=(8, 8))
plt.imshow(frame)
plt.axis(False)
plt.tight_layout()
plt.savefig("frame.png", dpi = 400)
```

```python
# divide each frame into 16 regions
# each frame is a square of 224x224
height, width = frame.shape[:-1]
num_regions = 16
area_of_each_region = (height * width) / num_regions
# since each region is also a square
side_of_region = int(sqrt(area_of_each_region))
```

```python
print(f"(Height, Width) -> {height}x{width}")
print("Num of regions to divide a frame into:", num_regions)
print("Area of each region:", area_of_each_region)
print("Side of each region:", side_of_region)
```

```python
# visualise the regions
figure, axes = plt.subplots(
    ncols=int(width // side_of_region),
    nrows=int(height // side_of_region),
    sharex=True,
    sharey=True,
    figsize=(8, 8),
)

for x, region_x in enumerate(range(0, height, int(side_of_region))):
    for y, region_y in enumerate(range(0, width, int(side_of_region))):
        frame_region = frame[
            region_x : region_x + int(side_of_region),
            region_y : region_y + int(side_of_region),
            :,
        ]
        axes[x][y].imshow(frame_region)
        axes[x][y].xaxis.set_label_position('top')
        axes[x][y].set_ylabel(x + 1)
        axes[x][y].set_xlabel(y + 1)
        axes[x][y].set_xticks([])
        axes[x][y].set_yticks([])
        axes[x][y].label_outer()
# figure.suptitle(f"cover -> Patchified")
plt.tight_layout()
plt.savefig("patchified_frame.png", dpi = 400)
plt.show()
```

```python
frame1_chunks = frame1.reshape(4, 56, 4, 56, 3).swapaxes(1, 2).reshape(-1, 56, 56, 3)
frame2_chunks = frame2.reshape(4, 56, 4, 56, 3).swapaxes(1, 2).reshape(-1, 56, 56, 3)
```

```python
# m_i and m_i+1 represents
# mean intensity values for a given colour channel for the region in two consecutive frames
m_1 = frame1_chunks.mean(axis=(1, 2))
m_2 = frame2_chunks.mean(axis=(1, 2))

# s_i and s_i+1 represents
# corresponding variance values for a given colour channel for the region in two consecutive frames
s_1 = frame1_chunks.var(axis=(1, 2))
s_2 = frame2_chunks.var(axis=(1, 2))
```

$$
\frac{\left[ \dfrac{s_i + s_{i+1}}{2} + \left( \dfrac{m_i - m_{i+1}}{2} \right)^2 \right]^2}{s_i \ast s_{i+1}}
$$

If $s_i = 0$ or $s_{i+1} = 0$, then set $s_i = s_{i+1} = 1$


```python
# compute the likelihood ratio across a single region and a single channel
a = (s_1 + s_2) / 2
b = ((m_1 - m_2) / 2) ** 2
c = (a + b) ** 2
d = s_1 * s_2
likelihood_ratio = c / d
```

```python
# using this sampling strategy to extract 'N' frames from a video
```

```python
def calculate_likelihood(frame1_chunks, frame2_chunks):
    """
    Applies the likelihood ratio formula from the notebook 
    between two sets of frame chunks.
    """
    # Calculate mean and variance 
    m_1 = frame1_chunks.mean(axis=(1, 2))
    m_2 = frame2_chunks.mean(axis=(1, 2))
    s_1 = frame1_chunks.var(axis=(1, 2))
    s_2 = frame2_chunks.var(axis=(1, 2))

    # Handle zero variance 
    # "If s_i = 0 or s_{i+1} = 0 then set s_i = s_{i+1} = 1"
    # We set any 0 variance to 1 to avoid division by zero.
    s_1[s_1 == 0] = 1
    s_2[s_2 == 0] = 1
    
    # Calculate likelihood ratio components [cite: 106, 107, 108, 109]
    a = (s_1 + s_2) / 2
    b = ((m_1 - m_2) / 2) ** 2
    c = (a + b) ** 2
    d = s_1 * s_2

    # Final likelihood ratio per region/channel [cite: 110]
    likelihood_ratio = c / d 
    
    # Average across the 3 color channels [cite: 113]
    avg_likelihood_per_region = likelihood_ratio.mean(axis=1)
    
    return avg_likelihood_per_region
```

```python
n_rows, n_cols = height // side_of_region, width // side_of_region  # 4, 4

frame_scores = []

N = 32
# Get the first frame
frame_prev = vr[0].asnumpy()
num_frames = len(vr)
for i in range(1, num_frames):
    frame_curr = vr[i].asnumpy()

    # Reshape frames into chunks
    frame1_chunks = (
        frame_prev.reshape(n_rows, side_of_region, n_cols, side_of_region, 3)
        .swapaxes(1, 2)
        .reshape(-1, side_of_region, side_of_region, 3)
    )
    frame2_chunks = (
        frame_curr.reshape(n_rows, side_of_region, n_cols, side_of_region, 3)
        .swapaxes(1, 2)
        .reshape(-1, side_of_region, side_of_region, 3)
    )

    # Calculate the 16 likelihood scores [cite: 113]
    likelihood_scores_per_region = calculate_likelihood(frame1_chunks, frame2_chunks)

    # Get a single score for the frame transition by averaging the region scores
    total_frame_change_score = likelihood_scores_per_region.mean()

    # Store the score and the index of the *current* frame
    frame_scores.append((total_frame_change_score, i))

    # Update frame_prev for the next iteration
    frame_prev = frame_curr

# Sort by score in descending order
sorted_frames = sorted(frame_scores, key=lambda x: x[0], reverse=True)

# Get the indices of the top N-1 frames with the most change
# We subtract 1 because we will always add frame 0 by default
top_indices = [idx for score, idx in sorted_frames[: N - 1]]

# Add frame 0 and sort the final indices
final_indices = sorted(list(set([0] + top_indices)))

# Retrieve the actual frames
selected_frames = vr.get_batch(final_indices).asnumpy()

print(f"Total frames: {num_frames}")
print(f"Requested: {N}")
print(f"Selected indices: {final_indices}")
```

```python
def plot_frames(frames, indices, title="Selected Keyframes"):
    """
    Plots a list of frames in a grid, using their indices as titles.
    Assumes 16 frames for a 4x4 grid.
    """
    if len(frames) != 16 or len(indices) != 16:
        print(f"Error: This function expects 16 frames, but got {len(frames)}.")
        print("Please adjust the 'nrows' and 'ncols' if you want to plot a different number.")
        return

    # Create a 4x4 grid of subplots
    fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))
    
    fig.suptitle(title, fontsize=16)
    
    for i, ax in enumerate(axes.flat):
        # axes.flat flattens the 4x4 grid into a 1D array
        
        # Plot the frame
        ax.imshow(frames[i])
        
        # Set the title of the subplot to the frame index
        ax.set_title(f"Frame: {indices[i]}")
        
        # Turn off the axes (x and y ticks)
        ax.axis("off")
        
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make room for suptitle
    plt.show()
```

```python
plot_frames(selected_frames[:16], final_indices[:16])
```
